```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
dados = pd.read_csv("vgsales.csv", sep = ";")
```

```python
sns.set(rc={'figure.figsize':(14, 8)})
sns.countplot(dados['PLataforma'])
```

```python
sns.set(rc={'figure.figsize':(14, 8)})
sns.countplot(dados['Gener'])
```

```python
sns.set(rc={'figure.figsize':(30, 8)})
sns.countplot(dados['Year'])
```

```python
sns.heatmap(dados.corr(), annot=True, fmt='.2f', linewidths=2)
```

```python
sns.distplot(dados['Year'])
```

```python
sns.boxplot(x='PLataforma', y='Year', data = dados)
```

```python
sns.set(rc={'figure.figsize':(14, 8)})
sns.jointplot(x='Global_Sales',y='Year',data=dados, kind='reg')
```

```python
fig, ax = plt.subplots(ncols=6, nrows=2, figsize=(20,10))
index = 0
ax = ax.flatten()

for col, value in dados.items():
    sns.boxplot(y=col, data=dados, color='r', ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```

```python
fig, ax = plt.subplots(ncols=6, nrows=2, figsize=(20,10))
index = 0
ax = ax.flatten()

for col, value in dados.items():
    sns.distplot(value, color='r', ax=ax[index])
    index += 1
plt.tight_layout(pad=0.5, w_pad=0.7, h_pad=5.0)
```
